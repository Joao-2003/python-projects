```
# # Avaliação N2 – Inteligência Artificial e Mineração de Dados
# **Professor:** Dr. Ricardo Marques de Carvalho
# **Curso:** Técnico em Administração – Eixo de Computação
# **Valor:** 10 pontos
# **Data de entrega:** (24-06-2025)
#
# ## Objetivo:
# - Aplicar algoritmos de classificação, regressão e clustering.
# - Simular um agente inteligente para recomendação baseada em padrões frequentes.
# - Analisar os resultados obtidos e refletir sobre sua aplicabilidade.


# ## Parte I – Classificação (3,0 pontos)
# **Cenário:** Uma empresa deseja automatizar a triagem de currículos.
#
# **1.** Implemente um modelo de **classificação com Árvores de Decisão**.
# **2.** Aplique outro classificador (ex: KNN ou Random Forest) e compare os resultados.
# **3.** Apresente uma **análise textual** baseada em precisão, recall e f1-score.


#%%
# Importando bibliotecas necessárias para a Parte I
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import numpy as np

# Carregando o dataset de currículos
try:
    curriculos = pd.read_csv('curriculos_ponto_virgula.csv', sep=';')
    print("Dataset de currículos carregado com sucesso!")
    display(curriculos.head())

    # --- Pré-processamento dos Dados ---

    # Separando features (X) e target (y)
    X = curriculos.drop('classe', axis=1)
    y = curriculos['classe']

    # Codificando a variável categórica 'escolaridade' com OneHotEncoder
    # Isso transforma a coluna de texto em colunas numéricas que o modelo pode entender
    one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    X_encoded = one_hot_encoder.fit_transform(X[['escolaridade']])

    # Criando um DataFrame com as colunas codificadas
    X_encoded_df = pd.DataFrame(X_encoded, columns=one_hot_encoder.get_feature_names_out(['escolaridade']))

    # Juntando as colunas codificadas com o restante das features numéricas
    X_final = pd.concat([X.drop('escolaridade', axis=1), X_encoded_df], axis=1)

    # Codificando a variável alvo 'classe' com LabelEncoder
    # Isso transforma as classes de texto ("Recusar", "Recomendar") em números (0, 1, 2...)
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    # Dividindo os dados em conjuntos de treino e teste (80% para treino, 20% para teste)
    X_train, X_test, y_train, y_test = train_test_split(X_final, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

    print("\nShape dos dados de treino:", X_train.shape)
    print("Shape dos dados de teste:", X_test.shape)

    # --- 1. Modelo com Árvore de Decisão ---
    print("\n--- Treinando Modelo: Árvore de Decisão ---")
    tree_clf = DecisionTreeClassifier(random_state=42)
    tree_clf.fit(X_train, y_train)
    y_pred_tree = tree_clf.predict(X_test)

    # --- 2. Modelo com Random Forest ---
    print("\n--- Treinando Modelo: Random Forest ---")
    forest_clf = RandomForestClassifier(random_state=42)
    forest_clf.fit(X_train, y_train)
    y_pred_forest = forest_clf.predict(X_test)

    # --- 3. Análise dos Resultados ---
    print("\n--- Análise Comparativa dos Classificadores ---\n")

    print("Relatório de Classificação - Árvore de Decisão:")
    print(classification_report(y_test, y_pred_tree, target_names=label_encoder.classes_, zero_division=0))

    print("Relatório de Classificação - Random Forest:")
    print(classification_report(y_test, y_pred_forest, target_names=label_encoder.classes_, zero_division=0))

except FileNotFoundError:
    print("Erro: O arquivo 'curriculos_ponto_virgula.csv' não foi encontrado.")
    print("Por favor, certifique-se de que o arquivo está no diretório correto.")


# ### Análise Textual da Classificação
#
# Comparamos dois modelos para a triagem de currículos: **Árvore de Decisão** e **Random Forest**.
#
# 1.  **Precisão (Precision):** Indica a proporção de predições positivas corretas.
#     * **Árvore de Decisão:** O modelo teve uma performance variável. Por exemplo, para a classe "Recomendar", a precisão foi de 86%, significando que, quando o modelo recomendou um candidato, ele estava certo em 86% das vezes.
#     * **Random Forest:** Mostrou uma precisão superior em geral, atingindo 95% para a classe "Recomendar". Isso indica uma maior confiabilidade nas suas recomendações positivas.
#
# 2.  **Recall (Revocação):** Mede a proporção de positivos reais que foram corretamente identificados.
#     * **Árvore de Decisão:** Para a classe "Recomendar", o recall foi de 95%, o que é excelente. Isso mostra que o modelo conseguiu identificar 95% de todos os candidatos que deveriam, de fato, ser recomendados.
#     * **Random Forest:** O recall para a mesma classe foi de 90%. Embora um pouco menor que o da Árvore de Decisão, ainda é um valor alto e demonstra eficácia em encontrar os candidatos relevantes.
#
# 3.  **F1-Score:** É a média harmônica entre precisão e recall, fornecendo uma métrica única que equilibra ambos.
#     * **Árvore de Decisão:** O F1-score geral foi de 91%, um resultado muito bom que indica um bom equilíbrio entre precisão e recall.
#     * **Random Forest:** Atingiu um F1-score geral de 93%, superando ligeiramente a Árvore de Decisão. Isso sugere que o Random Forest é um modelo mais robusto e equilibrado para esta tarefa.
#
# **Conclusão:** Ambos os modelos são eficazes, mas o **Random Forest apresentou um desempenho geral superior**, especialmente em sua precisão e F1-score. Em um cenário de triagem de currículos, onde tanto a identificação correta de bons candidatos (recall) quanto a confiabilidade das recomendações (precisão) são cruciais, o Random Forest seria a escolha mais segura e robusta.


# ## Parte II – Regressão (2,5 pontos)
# **Cenário:** Sistema que prevê o valor do aluguel de imóveis.
#
# **1.** Treine um modelo de **Regressão Linear**.
# **2.** Aplique **regressão com Lasso** e analise as variáveis eliminadas.
# **3.** Compare os dois modelos com base nos erros e no R².

#%%
# Importando bibliotecas necessárias para a Parte II
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Carregando o dataset de imóveis
try:
    imoveis = pd.read_csv('imoveis_ponto_virgula.csv', sep=';')
    print("Dataset de imóveis carregado com sucesso!")
    display(imoveis.head())

    # --- Pré-processamento dos Dados ---
    X = imoveis.drop('preco_aluguel', axis=1)
    y = imoveis['preco_aluguel']

    # Codificando a variável 'localizacao'
    one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    X_encoded = one_hot_encoder.fit_transform(X[['localizacao']])
    X_encoded_df = pd.DataFrame(X_encoded, columns=one_hot_encoder.get_feature_names_out(['localizacao']))
    X_final = pd.concat([X.drop('localizacao', axis=1), X_encoded_df], axis=1)

    # Dividindo os dados
    X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

    # --- 1. Modelo de Regressão Linear ---
    print("\n--- Treinando Modelo: Regressão Linear ---")
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, y_train)
    y_pred_lin = lin_reg.predict(X_test)

    # --- 2. Modelo de Regressão com Lasso ---
    print("\n--- Treinando Modelo: Regressão Lasso ---")
    lasso_reg = Lasso(alpha=1.0, random_state=42) # alpha é o parâmetro de regularização
    lasso_reg.fit(X_train, y_train)
    y_pred_lasso = lasso_reg.predict(X_test)

    # --- Análise das variáveis eliminadas pelo Lasso ---
    print("\n--- Análise de Coeficientes do Lasso ---")
    lasso_coeffs = pd.DataFrame({'Variável': X_final.columns, 'Coeficiente Lasso': lasso_reg.coef_})
    print("Coeficientes do modelo Lasso:")
    display(lasso_coeffs)

    variaveis_eliminadas = lasso_coeffs[lasso_coeffs['Coeficiente Lasso'] == 0]['Variável'].tolist()
    if variaveis_eliminadas:
        print(f"\nVariáveis eliminadas pelo Lasso (coeficiente = 0): {variaveis_eliminadas}")
    else:
        print("\nNenhuma variável foi completamente eliminada pelo Lasso com o alpha atual.")

    # --- 3. Comparação dos Modelos ---
    print("\n--- Análise Comparativa dos Modelos de Regressão ---\n")

    # Métricas para Regressão Linear
    mse_lin = mean_squared_error(y_test, y_pred_lin)
    r2_lin = r2_score(y_test, y_pred_lin)

    # Métricas para Regressão Lasso
    mse_lasso = mean_squared_error(y_test, y_pred_lasso)
    r2_lasso = r2_score(y_test, y_pred_lasso)

    print("Métricas - Regressão Linear:")
    print(f"Erro Quadrático Médio (MSE): {mse_lin:.2f}")
    print(f"Coeficiente de Determinação (R²): {r2_lin:.4f}")

    print("\nMétricas - Regressão Lasso:")
    print(f"Erro Quadrático Médio (MSE): {mse_lasso:.2f}")
    print(f"Coeficiente de Determinação (R²): {r2_lasso:.4f}")

except FileNotFoundError:
    print("Erro: O arquivo 'imoveis_ponto_virgula.csv' não foi encontrado.")
    print("Por favor, certifique-se de que o arquivo está no diretório correto.")


# ### Análise Textual da Regressão
#
# Comparamos dois modelos para prever o preço do aluguel: **Regressão Linear** e **Regressão com Lasso**.
#
# 1.  **Análise das Variáveis (Lasso):** A Regressão Lasso é útil por realizar "seleção de features", zerando os coeficientes das variáveis menos importantes. No nosso caso, com o `alpha` padrão de 1.0, o Lasso **não eliminou completamente nenhuma variável**. Ele apenas reduziu a magnitude dos coeficientes, indicando que todas as variáveis (`quartos`, `metragem`, `idade_imovel` e as localizações) têm alguma relevância para o modelo. Para eliminar variáveis, seria necessário aumentar o valor de `alpha`.
#
# 2.  **Erro Quadrático Médio (MSE):** Mede a média dos erros ao quadrado. Quanto menor, melhor.
#     * **Regressão Linear:** Apresentou um MSE de `3430.70`.
#     * **Regressão Lasso:** Teve um MSE ligeiramente menor, de `3429.87`.
#     * A diferença é mínima, mas tecnicamente o Lasso teve um desempenho um pouco melhor na redução do erro.
#
# 3.  **Coeficiente de Determinação (R²):** Indica a proporção da variância da variável dependente que é previsível a partir das variáveis independentes. Varia de 0 a 1, e quanto mais próximo de 1, melhor o modelo explica os dados.
#     * **Regressão Linear:** Obteve um R² de `0.9980`.
#     * **Regressão Lasso:** Obteve um R² de `0.9980`.
#     * Ambos os modelos alcançaram um R² extremamente alto e praticamente idêntico, indicando que eles explicam mais de 99.8% da variação nos preços dos aluguéis. Isso sugere uma forte relação linear nos dados.
#
# **Conclusão:** Ambos os modelos foram extraordinariamente eficazes, com um poder de predição quase perfeito (R² > 0.99). A **Regressão Lasso** teve uma leve vantagem em termos de erro (MSE), e sua capacidade de simplificar o modelo (mesmo que não tenha zerado coeficientes neste caso) o torna uma escolha marginalmente preferível, pois tende a ser mais robusto contra a superestimação (overfitting). Na prática, para este dataset, os dois são praticamente intercambiáveis e altamente precisos.


# ## Parte III – Clustering (2,0 pontos)
# **Cenário:** Supermercado deseja agrupar clientes por perfil de compra.
#
# **1.** Aplique o **K-Means** com 3 grupos.
# **2.** Descreva ações possíveis para cada grupo com base nos dados.

#%%
# Importando bibliotecas necessárias para a Parte III
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando o dataset de clientes
try:
    clientes = pd.read_csv('clientes_ponto_virgula.csv', sep=';')
    print("Dataset de clientes carregado com sucesso!")
    display(clientes.head())

    # --- Pré-processamento dos Dados ---
    # É crucial escalar os dados para o K-Means, pois ele é sensível à escala das variáveis.
    scaler = StandardScaler()
    clientes_scaled = scaler.fit_transform(clientes)
    clientes_scaled_df = pd.DataFrame(clientes_scaled, columns=clientes.columns)

    # --- 1. Aplicação do K-Means ---
    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    kmeans.fit(clientes_scaled_df)

    # Adicionando os rótulos dos clusters ao DataFrame original
    clientes['cluster'] = kmeans.labels_

    print("\nDistribuição dos clientes por cluster:")
    print(clientes['cluster'].value_counts())

    # --- Análise dos Clusters ---
    # Calculando a média das características para cada cluster
    cluster_analysis = clientes.groupby('cluster').mean()
    print("\nAnálise das características médias por cluster:")
    display(cluster_analysis)

    # --- Visualização dos Clusters ---
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=clientes, x='gasto_anual', y='frequencia_compras', hue='cluster', palette='viridis', s=100)
    plt.title('Clusters de Clientes por Gasto Anual vs. Frequência de Compras')
    plt.xlabel('Gasto Anual (R$)')
    plt.ylabel('Frequência de Compras (Anual)')
    plt.legend(title='Cluster')
    plt.grid(True)
    plt.show()

except FileNotFoundError:
    print("Erro: O arquivo 'clientes_ponto_virgula.csv' não foi encontrado.")
    print("Por favor, certifique-se de que o arquivo está no diretório correto.")


# ### 2. Análise e Ações Sugeridas para Cada Grupo
#
# Com base na análise das características médias de cada cluster, podemos definir os seguintes perfis e ações:
#
# #### **Grupo 0: Clientes de Baixo Valor e Frequência Média**
# * **Perfil:** Este grupo tem o **menor gasto anual** (`R$ 3.864`) e o **menor valor médio** por compra (`R$ 178`). A frequência de compras é intermediária. São clientes que compram o essencial, com baixo ticket médio.
# * **Ações Possíveis:**
#     * **Promoções de Volume:** Oferecer descontos progressivos ("Leve 3, Pague 2") para incentivar um aumento no volume de itens por compra.
#     * **Programas de Fidelidade por Ponto:** Criar um sistema onde cada real gasto gera pontos que podem ser trocados por produtos básicos, incentivando-os a concentrar suas compras no supermercado.
#     * **Marketing Direcionado:** Enviar ofertas de produtos de baixo custo e de necessidade básica.
#
# #### **Grupo 1: Clientes de Alto Valor (VIPs)**
# * **Perfil:** Este é o grupo mais valioso. Possuem de longe o **maior gasto anual** (`R$ 13.064`) e o **maior valor médio** por compra (`R$ 418`). Embora não sejam os que compram com maior frequência, cada visita ao mercado resulta em um gasto significativo.
# * **Ações Possíveis:**
#     * **Programa de Fidelidade Exclusivo:** Oferecer benefícios premium, como cashback, descontos exclusivos em produtos de alto valor agregado (vinhos, carnes nobres, importados) e atendimento personalizado.
#     * **Eventos de Relacionamento:** Convidá-los para degustações, workshops de culinária ou eventos de lançamento de produtos.
#     * **Upselling:** Treinar a equipe para sugerir produtos complementares ou de qualidade superior durante a compra.
#
# #### **Grupo 2: Clientes Fieis e Frequentes**
# * **Perfil:** Este grupo se destaca pela **altíssima frequência de compras** (média de 29 visitas anuais). O gasto anual (`R$ 7.078`) e o valor médio (`R$ 240`) são moderados. São os clientes do dia a dia, que visitam o supermercado constantemente.
# * **Ações Possíveis:**
#     * **Clube de Assinatura:** Oferecer um serviço de assinatura para entrega de produtos recorrentes (hortifrúti, pães, leite) com um pequeno desconto, garantindo a recorrência e conveniência.
#     * **Promoções Relâmpago:** Enviar notificações via app com ofertas de curta duração para produtos que eles costumam comprar, incentivando mais uma visita à loja.
#     * **Gamificação:** Criar desafios no app do supermercado ("Visite-nos 5 vezes este mês e ganhe um brinde"), capitalizando em sua alta frequência.


# ## Parte IV – Agente Inteligente com Regras de Associação (2,5 pontos)
# **Cenário:** Sistema que recomenda produtos com base em padrões de compra.
#
# **1.** Extraia padrões frequentes com Apriori ou FP-Growth.
# **2.** Simule um **agente** que recomenda um item adicional com base nas regras aprendidas.

#%%
# Importando bibliotecas necessárias para a Parte IV
import pandas as pd
import ast
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Carregando o dataset de transações
try:
    transacoes_df = pd.read_csv('transacoes_ponto_virgula.csv', sep=';')
    # Convertendo a string de lista para uma lista de verdade
    transacoes_df['transacao'] = transacoes_df['transacao'].apply(ast.literal_eval)
    transacoes = list(transacoes_df['transacao'])
    print("Dataset de transações carregado com sucesso!")
    display(transacoes_df.head())

    # --- 1. Extração de Padrões Frequentes com Apriori ---

    # Preparando os dados para o algoritmo Apriori
    te = TransactionEncoder()
    te_ary = te.fit(transacoes).transform(transacoes)
    df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

    # Aplicando o Apriori para encontrar itemsets frequentes
    # min_support=0.05 significa que o itemset deve aparecer em pelo menos 5% das transações
    frequent_itemsets = apriori(df_encoded, min_support=0.05, use_colnames=True)
    print("\n--- Itemsets Frequentes (Suporte > 5%) ---")
    display(frequent_itemsets.sort_values(by='support', ascending=False))

    # Gerando as regras de associação a partir dos itemsets frequentes
    # min_threshold=0.5 para a métrica 'confidence'
    regras = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
    print("\n--- Regras de Associação (Confiança > 50%) ---")
    # Exibindo as regras mais relevantes (maior confiança e lift)
    display(regras.sort_values(by=['confidence', 'lift'], ascending=[False, False]))

    # --- 2. Simulação do Agente de Recomendação ---

    def agente_recomendador(carrinho, regras_df):
        """
        Simula um agente que recomenda um item com base no carrinho de compras atual.

        Args:
            carrinho (set): Um conjunto de itens no carrinho do cliente.
            regras_df (pd.DataFrame): DataFrame com as regras de associação.

        Returns:
            str: A recomendação de um item ou uma mensagem padrão.
        """
        recomendacoes_possiveis = []

        # Itera sobre as regras para encontrar correspondências
        for _, regra in regras_df.iterrows():
            antecedentes = set(regra['antecedents'])

            # Se os itens no carrinho contêm os itens de uma regra
            if antecedentes.issubset(carrinho):
                consequente = list(regra['consequents'])[0]
                # Verifica se o item recomendado já não está no carrinho
                if consequente not in carrinho:
                    recomendacoes_possiveis.append((consequente, regra['confidence'], regra['lift']))

        # Se encontrou recomendações, ordena pela maior confiança e depois pelo maior lift
        if recomendacoes_possiveis:
            recomendacoes_possiveis.sort(key=lambda x: (x[1], x[2]), reverse=True)
            melhor_recomendacao = recomendacoes_possiveis[0][0]
            return f"Vimos que você está levando {', '.join(carrinho)}. Que tal adicionar **{melhor_recomendacao}** também?"
        else:
            return "Aproveite nossas ofertas do dia na seção de hortifrúti!"

    # --- Testando o Agente ---
    print("\n--- Simulação do Agente Inteligente de Recomendação ---\n")

    # Cenário 1: Cliente coloca pão no carrinho
    carrinho_1 = {'pão'}
    print(f"Carrinho do Cliente: {carrinho_1}")
    print(f"Recomendação do Agente: {agente_recomendador(carrinho_1, regras)}")

    print("-" * 30)

    # Cenário 2: Cliente coloca manteiga no carrinho
    carrinho_2 = {'manteiga'}
    print(f"Carrinho do Cliente: {carrinho_2}")
    print(f"Recomendação do Agente: {agente_recomendador(carrinho_2, regras)}")

    print("-" * 30)

    # Cenário 3: Cliente coloca feijão e arroz
    carrinho_3 = {'feijão', 'arroz'}
    print(f"Carrinho do Cliente: {carrinho_3}")
    print(f"Recomendação do Agente: {agente_recomendador(carrinho_3, regras)}")

except FileNotFoundError:
    print("Erro: O arquivo 'transacoes_ponto_virgula.csv' não foi encontrado.")
    print("Por favor, certifique-se de que o arquivo está no diretório correto.")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")


# ## Parte V – Síntese Reflexiva (0,5 ponto)
# Responda brevemente:
# - Qual modelo foi mais útil e por quê?
# - Quais dificuldades você teve ao aplicar os modelos?


# ### Resposta Reflexiva
#
# **1. Qual modelo foi mais útil e por quê?**
#
# O modelo mais útil, considerando o impacto direto e a clareza dos insights para o negócio, foi o de **Clustering com K-Means (Parte III)**. Embora os modelos de classificação e regressão tenham apresentado altíssima precisão técnica, a segmentação de clientes gera insights acionáveis de forma muito intuitiva. Ser capaz de dividir a base de clientes em grupos distintos como "Clientes VIP", "Clientes Fiéis" e "Clientes de Baixo Valor" permite que uma empresa (neste caso, o supermercado) crie estratégias de marketing, vendas e fidelização extremamente direcionadas e eficientes. A capacidade de traduzir os clusters em personas de clientes e planejar ações específicas para cada um deles oferece um valor estratégico imenso e fácil de comunicar para as equipes de negócio.
#
# **2. Quais dificuldades você teve ao aplicar os modelos?**
#
# As principais dificuldades encontradas foram:
#
# * **Pré-processamento de Dados:** A necessidade de tratar variáveis categóricas foi um ponto crucial em quase todos os modelos. Decidir entre `LabelEncoder` e `OneHotEncoder` e aplicá-los corretamente, especialmente antes de dividir os dados em treino e teste, exige atenção para evitar vazamento de dados (data leakage). Nos modelos de Regressão e Classificação, essa etapa foi fundamental para que os algoritmos pudessem processar as informações.
#
# * **Interpretação e Parametrização:** No modelo de Regras de Associação (Parte IV), a escolha dos parâmetros `min_support` e `min_confidence` é um desafio. Um suporte muito alto pode não gerar nenhuma regra, enquanto um suporte muito baixo pode gerar milhares de regras irrelevantes. Foi preciso um ajuste fino para encontrar um equilíbrio que gerasse insights úteis e acionáveis. Da mesma forma, no K-Means, a definição do número de clusters (k=3) foi dada, mas em um cenário real, determinar o 'k' ótimo (usando métodos como o do cotovelo) seria uma etapa adicional e, por vezes, subjetiva.

```
